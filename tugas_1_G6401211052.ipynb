```python
#4.12
import random

def tortoise_move():
    move_type = random.randint(1, 10)
    if 1 <= move_type <= 5:
        return 3  # Fast plod
    elif 6 <= move_type <= 7:
        return -6  # Slip
    else:
        return 1  # Slow plod

def hare_move():
    move_type = random.randint(1, 10)
    if 1 <= move_type <= 2:
        return 0  # Sleep
    elif 3 <= move_type <= 4:
        return 9  # Big hop
    elif 5 <= move_type <= 5:
        return -12  # Big slip
    elif 6 <= move_type <= 10:
        return 1  # Small hop
    else:
        return -2  # Small slip

def display_race(tortoise, hare):
    for i in range(1, 71):
        if i == tortoise and i == hare:
            print("OUCH!!!", end='')
        elif i == tortoise:
            print("T", end='')
        elif i == hare:
            print("H", end='')
        else:
            print(" ", end='')
    print()

def race():
    tortoise_position = 1
    hare_position = 1

    print("BANG !!!!!")
    print("AND THEY'RE OFF !!!!!")

    while True:
        tortoise_position += tortoise_move()
        hare_position += hare_move()

        if tortoise_position < 1:
            tortoise_position = 1
        if hare_position < 1:
            hare_position = 1

        display_race(tortoise_position, hare_position)

        if tortoise_position >= 70 and hare_position >= 70:
            print("It's a tie!")
            break
        elif tortoise_position >= 70:
            print("TORTOISE WINS!!! YAY!!!")
            break
        elif hare_position >= 70:
            print("Hare wins. Yuch.")
            break

if __name__ == "__main__":
    race()
```

```python
#4.17
import random

def generate_question(difficulty_level, problem_type):
    if problem_type == 1:  # Addition
        num1 = random.randint(1, 9 if difficulty_level == 1 else 99)
        num2 = random.randint(1, 9 if difficulty_level == 1 else 99)
        operator = '+'
    elif problem_type == 2:  # Subtraction
        num1 = random.randint(1, 9 if difficulty_level == 1 else 99)
        num2 = random.randint(1, 9 if difficulty_level == 1 else 99)
        num1, num2 = max(num1, num2), min(num1, num2)
        operator = '-'
    elif problem_type == 3:  # Multiplication
        num1 = random.randint(1, 9 if difficulty_level == 1 else 99)
        num2 = random.randint(1, 9 if difficulty_level == 1 else 99)
        operator = 'x'
    elif problem_type == 4:  # Division
        num2 = random.randint(1, 9 if difficulty_level == 1 else 99)
        num1 = num2 * random.randint(1, 9 if difficulty_level == 1 else 99)
        operator = '/'
    elif problem_type == 5:  # Random mixture
        sub_problem_type = random.randint(1, 4)
        return generate_question(difficulty_level, sub_problem_type)

    return num1, operator, num2

def get_response(correct):
    responses = {
        'correct': ['Very good!', 'Nice work!', 'Keep up the good work!'],
        'incorrect': ['No. Please try again.', 'Wrong. Try once more.', 'No. Keep trying.']
    }
    return random.choice(responses['correct' if correct else 'incorrect'])

def ask_question():
    while True:
        difficulty_level = int(input("Enter the difficulty level (1 or 2): "))
        if difficulty_level not in [1, 2]:
            print("Invalid difficulty level. Use 1 or 2.")
            continue

        problem_type = int(input("Choose a type of arithmetic problem (1: Addition, 2: Subtraction, 3: Multiplication, 4: Division, 5: Random mixture): "))
        if problem_type not in [1, 2, 3, 4, 5]:
            print("Invalid problem type. Use 1, 2, 3, 4, or 5.")
            continue
        
        num1, operator, num2 = generate_question(difficulty_level, problem_type)
        correct_answer = eval(f"{num1} {operator} {num2}")
        while True:
            user_answer = float(input(f"How much is {num1} {operator} {num2}? "))
            is_correct = user_answer == correct_answer
            print(get_response(is_correct))
            if is_correct:
                break

if __name__ == "__main__":
    print("Welcome to the Arithmetic Quiz!")
    ask_question()
```

```python
############################# exercise 5 #############################
#5.2
def mystery(sequence):
    return sequence == sorted(sequence)
#Fungsi mystery menerima satu argumen, yang berupa sebuah data list atau tuple, dimana fungsi tersebut akan mengembalikan True jika elemen-elemen dalam urutan tersebut sudah terurut dalam urutan naik (Ascending) dan False jika tidak.

# Fungsi ini bekerja dengan membandingkan data masukan dengan data versi yang telah terurut menggunakan fungsi sorted. Jika data masukan sama dengan data versi terurutnya, itu berarti elemen-elemennya sudah dalam urutan membesar, dan fungsi akan mengembalikan True. Sebaliknya, jika urutan masukan tidak sama dengan versi urutannya, maka akan mengembalikan False.

# Berikut adalah contoh bagaimana fungsi ini bekerja:
sequence1 = [1, 2, 3, 4, 5]
result1 = mystery(sequence1)  # akan mengembalikan True karena elemennya sudah terurut.
print(result1)

sequence2 = [5, 4, 3, 2, 1]
result2 = mystery(sequence2)  # akan mengembalikan False karena elemennya tidak terurut dalam urutan naik.
print(result2)
#Jadi, fungsi mystery digunakan untuk memeriksa apakah urutan tertentu sudah terurut dalam urutan membesar (Ascending) atau tidak.
```

```python
#5.8
def sieve_of_eratosthenes():
    n = 1000  # Batas atas untuk pencarian angka prima
    primes = [True] * (n + 1)  # Inisialisasi semua elemen dengan True
    primes[0] = primes[1] = False  # Anggap 0 dan 1 bukan angka prima

    # Menandai angka-angka yang bukan prima
    for p in range(2, int(n**0.5) + 1):
        if primes[p]:
            for i in range(p * p, n + 1, p):
                primes[i] = False

    # Menampilkan angka-angka prima yang ditemukan
    prime_numbers = [i for i, is_prime in enumerate(primes) if is_prime]
    return prime_numbers

# Memanggil fungsi Sieve of Eratosthenes
prime_numbers = sieve_of_eratosthenes()

# Menampilkan angka prima kurang dari 1000
print("Angka prima kurang dari 1000:")
for prime in prime_numbers:
    print(prime, end=" ")
```

```python
#5.9
def is_palindrome(string):
    # Menghapus spasi dan tanda baca, dan mengubah huruf menjadi huruf kecil
    cleaned_string = ''.join(char.lower() for char in string if char.isalnum())

    # Membalikkan string dan membandingkannya dengan string asli
    return cleaned_string == cleaned_string[::-1]

# Contoh penggunaan fungsi
string1 = "A man, a plan, a canal, Panama"
string2 = "racecar"
string3 = "Hello, World!"

result1 = is_palindrome(string1)
result2 = is_palindrome(string2)
result3 = is_palindrome(string3)

print(f"'{string1}' adalah palindrom: {result1}")
print(f"'{string2}' adalah palindrom: {result2}")
print(f"'{string3}' adalah palindrom: {result3}")
```

```python
#5.10
def generate_anagrams(string):
    # Base case: jika panjang string adalah 1, kembalikan string itu sendiri sebagai anagram tunggal
    if len(string) == 1:
        return [string]

    # List untuk menyimpan semua anagram
    anagrams = []

    # Iterasi melalui setiap karakter dalam string
    for char in string:
        # Mengambil karakter yang akan dihapus dari string
        remaining_chars = string.replace(char, "", 1)
        
        # Membuat anagram dari karakter yang diambil
        partial_anagrams = generate_anagrams(remaining_chars)
        
        # Menambahkan karakter yang diambil ke setiap anagram parsial
        anagrams.extend([char + partial_anagram for partial_anagram in partial_anagrams])

    return anagrams

# Contoh penggunaan skrip
input_string = "tes"
all_anagrams = generate_anagrams(input_string)

# Menampilkan semua anagram
for anagram in all_anagrams:
    print(anagram)
```

```python
#5.11
def summarize_letters(input_string):
    # Menghapus spasi dan tanda baca, dan mengubah huruf menjadi huruf kecil
    cleaned_string = ''.join(char.lower() for char in input_string if char.isalnum())

    # Menghitung frekuensi setiap huruf dalam string
    letter_counts = {}
    for char in cleaned_string:
        if char in letter_counts:
            letter_counts[char] += 1
        else:
            letter_counts[char] = 1

    # Menghasilkan daftar tupel berisi huruf unik dan frekuensinya
    letter_summary = list(letter_counts.items())
    letter_summary.sort()

    return letter_summary

# Contoh penggunaan fungsi
input_string = "Hello, World!"
letter_summary = summarize_letters(input_string)

# Menampilkan setiap huruf dengan frekuensinya
for letter, count in letter_summary:
    print(f"{letter}: {count}")

# Mengecek apakah string memiliki semua huruf alfabet
is_full_alphabet = set("abcdefghijklmnopqrstuvwxyz").issubset([letter for letter, count in letter_summary])
if is_full_alphabet:
    print("String memiliki semua huruf alfabet.")
else:
    print("String tidak memiliki semua huruf alfabet.")
```

```python
#5.12
import itertools

# Tabel digit-huruf
digit_to_letters = {
    '2': 'ABC',
    '3': 'DEF',
    '4': 'GHI',
    '5': 'JKL',
    '6': 'MNO',
    '7': 'PRS',
    '8': 'TUV',
    '9': 'WXY'
}

def generate_word_combinations(phone_number):
    # Menghapus digit 0 dan 1 dari nomor telepon
    phone_number = ''.join(char for char in phone_number if char not in '01')

    # Menghasilkan semua kombinasi kata berjumlah tujuh huruf sesuai dengan nomor telepon
    combinations = [''.join(combination) for combination in itertools.product(*[digit_to_letters[digit] for digit in phone_number])]
    return combinations

# Contoh penggunaan skrip
phone_number = "7382273"
word_combinations = generate_word_combinations(phone_number)

# Menampilkan semua kombinasi kata
for word in word_combinations:
    print(word)
```

```python
#5.14
def is_ordered(sequence):
    if all(sequence[i] <= sequence[i + 1] for i in range(len(sequence) - 1)):
        return True
    return False

# Contoh pengujian dengan daftar (list)
sorted_list = [1, 2, 3, 4, 5]
unsorted_list = [4, 2, 6, 1, 5]

print("Sorted List (List):", is_ordered(sorted_list))  # Output: True
print("Unsorted List (List):", is_ordered(unsorted_list))  # Output: False

# Contoh pengujian dengan tupel (tuple)
sorted_tuple = (1, 2, 3, 4, 5)
unsorted_tuple = (4, 2, 6, 1, 5)

print("Sorted Tuple (Tuple):", is_ordered(sorted_tuple))  # Output: True
print("Unsorted Tuple (Tuple):", is_ordered(unsorted_tuple))  # Output: False

# Contoh pengujian dengan string
sorted_string = "abcde"
unsorted_string = "edcba"

print("Sorted String (String):", is_ordered(sorted_string))  # Output: True
print("Unsorted String (String):", is_ordered(unsorted_string))  # Output: False
```

```python
#5.15
from operator import itemgetter

# Data contoh faktur toko peralatan sebagai tuple
invoices = [
    ('83', 'Electric sander', 7, 57.98),
    ('24', 'Power saw', 18, 99.99),
    ('7', 'Sledge hammer', 11, 21.50),
    ('77', 'Hammer', 76, 11.99),
    ('39', 'Jig saw', 3, 79.50)
]

# a) Mengurutkan faktur berdasarkan deskripsi barang
sorted_by_description = sorted(invoices, key=itemgetter(1))
print("a) Diurutkan berdasarkan Deskripsi Barang:")
for invoice in sorted_by_description:
    print(invoice)

# b) Mengurutkan faktur berdasarkan harga
sorted_by_price = sorted(invoices, key=itemgetter(3))
print("\nb) Diurutkan berdasarkan Harga:")
for invoice in sorted_by_price:
    print(invoice)

# c) Memetakan setiap faktur ke tuple yang berisi deskripsi barang dan kuantitas, lalu diurutkan berdasarkan kuantitas
mapped_and_sorted_by_quantity = sorted([(invoice[1], invoice[2]) for invoice in invoices], key=itemgetter(1))
print("\nc) Diurutkan berdasarkan Kuantitas:")
for invoice in mapped_and_sorted_by_quantity:
    print(invoice)

# d) Memetakan setiap faktur ke tuple yang berisi deskripsi barang dan nilai faktur (hasil perkalian kuantitas dan harga barang), lalu diurutkan berdasarkan nilai faktur
mapped_and_sorted_by_value = sorted([(invoice[1], invoice[2] * invoice[3]) for invoice in invoices], key=itemgetter(1))
print("\nd) Diurutkan berdasarkan Nilai Faktur:")
for invoice in mapped_and_sorted_by_value:
    print(invoice)

# e) Menyaring hasilnya untuk nilai faktur dalam kisaran $200 hingga $500
filtered_invoices = [(invoice[1], invoice[2] * invoice[3]) for invoice in invoices if 200 <= invoice[2] * invoice[3] <= 500]
print("\ne) Faktur dalam Rentang $200 hingga $500:")
for invoice in filtered_invoices:
    print(invoice)

# f) Menghitung total dari semua faktur
total_nilai_faktur = sum(invoice[2] * invoice[3] for invoice in invoices)
print("\nf) Total Nilai Faktur:", total_nilai_faktur)
```

```python
#5.17
def is_odd(x):
    print(f"Called filter with {x}")
    return x % 2 != 0

def square(x):
    print(f"Called map with {x}")
    return x ** 2

numbers = [10, 3, 7, 1, 9, 4, 2, 8, 5, 6]
#output
result = list(map(square, filter(is_odd, numbers)))
print(result)
```

```python
############################# exercise 6 #############################
#6.5
def count_duplicate_words(sentence):
    # Mengubah semua huruf menjadi huruf kecil dan memisahkan kalimat menjadi kata-kata
    words = sentence.lower().split()

    # Membuat kamus untuk menghitung jumlah kata
    word_counts = {}

    # Menghitung jumlah kemunculan setiap kata dalam kalimat
    for word in words:
        if word in word_counts:
            word_counts[word] += 1
        else:
            word_counts[word] = 1

    # Mencetak kata-kata yang memiliki lebih dari satu kemunculan (duplikat)
    print("Duplicate Words:")
    for word, count in word_counts.items():
        if count > 1:
            print(f"{word}: {count}")

# Contoh penggunaan skrip
sentence = "This is a sample sentence. This is another sample sentence with sample words."

count_duplicate_words(sentence)
```

```python
#6.6
import string

def get_unique_words(word_list):
    # Menggabungkan semua kata menjadi satu string, mengubah huruf menjadi huruf kecil, dan memisahkan kata-kata
    words = ' '.join(word_list).lower().split()

    # Menghapus tanda baca dari kata-kata
    words = [''.join(char for char in word if char not in string.punctuation) for word in words]

    # Menggunakan set untuk mendapatkan kata-kata unik
    unique_words = set(words)

    # Mengurutkan kata-kata unik dalam urutan alfabet
    unique_words_sorted = sorted(unique_words)

    return unique_words_sorted

# Contoh penggunaan fungsi
sentences = [
    "This is a sample sentence.",
    "This is another sample sentence with more words.",
    "These sentences contain duplicate words, sentences have words."
]

for sentence in sentences:
    word_list = sentence.split()
    unique_words = get_unique_words(word_list)
    print(f"- Kata unik dari '{sentence}': {', '.join(unique_words)}")
```

```python
#6.7
# Meminta pengguna memasukkan sebuah kalimat
sentence = input("Masukkan sebuah kalimat: ")

# Menginisialisasi kamus untuk menghitung jumlah kemunculan huruf
letter_counts = {}

# Menghitung jumlah kemunculan setiap huruf dalam kalimat
for char in sentence:
    if char.isalpha():  # Hanya memproses huruf (mengabaikan spasi dan karakter lainnya)
        char = char.lower()  # Mengabaikan huruf besar/kecil dengan mengubah menjadi huruf kecil
        if char in letter_counts:
            letter_counts[char] += 1
        else:
            letter_counts[char] = 1

# Menampilkan tabel huruf dan jumlah kemunculannya dalam urutan terurut
print("Huruf".ljust(6)+"Jumlah Kemunculan".rjust(20))
for char in sorted(letter_counts.keys()):
    print(char.ljust(6)+letter_counts[char].__str__().center(20))

# Menemukan huruf-huruf yang tidak muncul dalam kalimat
alphabet = set("abcdefghijklmnopqrstuvwxyz")
missing_letters = alphabet - set(letter_counts.keys())

# Menampilkan huruf-huruf yang tidak muncul dalam kalimat
print("Huruf yang tidak muncul dalam kalimat:", ', '.join(sorted(missing_letters)))
```

```python
#6.8
# Kamus yang menghubungkan angka dengan kata-kata
number_words = {
    0: 'ZERO',
    1: 'ONE',
    2: 'TWO',
    3: 'THREE',
    4: 'FOUR',
    5: 'FIVE',
    6: 'SIX',
    7: 'SEVEN',
    8: 'EIGHT',
    9: 'NINE',
    10: 'TEN',
    11: 'ELEVEN',
    12: 'TWELVE',
    13: 'THIRTEEN',
    14: 'FOURTEEN',
    15: 'FIFTEEN',
    16: 'SIXTEEN',
    17: 'SEVENTEEN',
    18: 'EIGHTEEN',
    19: 'NINETEEN',
    20: 'TWENTY',
    30: 'THIRTY',
    40: 'FORTY',
    50: 'FIFTY',
    60: 'SIXTY',
    70: 'SEVENTY',
    80: 'EIGHTY',
    90: 'NINETY'
}

# Meminta pengguna memasukkan jumlah numerik
amount = float(input("Masukkan jumlah numerik (kurang dari 1000): "))

# Fungsi untuk mengonversi dua digit terakhir menjadi kata-kata
def convert_two_digits(number):
    if number in number_words:
        return number_words[number]
    else:
        tens = number // 10 * 10
        ones = number % 10
        return number_words[tens] + ' ' + number_words[ones]

# Mengonversi jumlah numerik menjadi kata-kata
def convert_to_words(amount):
    if amount < 1:
        return "ZERO"
    
    integer_part = int(amount)
    decimal_part = int((amount - integer_part) * 100)

    words = []
    
    if integer_part >= 100:
        hundreds = integer_part // 100
        words.append(number_words[hundreds] + ' HUNDRED')
        integer_part %= 100
    
    if integer_part >= 20:
        tens = integer_part // 10 * 10
        words.append(number_words[tens])
        integer_part %= 10

    if integer_part > 0:
        words.append(number_words[integer_part])

    words.append('AND')
    words.append(f'{decimal_part:02d}/100')

    return ' '.join(words)

# Mengonversi dan menampilkan jumlah numerik dalam kata-kata
amount_in_words = convert_to_words(amount)
print("Jumlah dalam kata-kata:", amount_in_words)
```

```python
#6.9
# Kamus yang menghubungkan negara dengan top-level domains (TLDs)
tlds = {'Canada': 'ca', 'United States': 'us', 'Mexico': 'mx'}

# a) Mengecek apakah kamus mengandung kunci 'Canada'
contains_canada = 'Canada' in tlds
print(f'a) Kamus mengandung kunci "Canada": {contains_canada}')

# b) Mengecek apakah kamus mengandung kunci 'France'
contains_france = 'France' in tlds
print(f'b) Kamus mengandung kunci "France": {contains_france}')

# c) Iterasi melalui pasangan kunci-nilai dan menampilkannya dalam format dua kolom
print('c) Isi kamus (kunci - nilai):')
for country, tld in tlds.items():
    print(f'{country:<15} - {tld}')

# d) Menambahkan pasangan kunci-nilai 'Sweden' dan 'sw' (yang salah)
tlds['Sweden'] = 'sw'

# e) Memperbarui nilai untuk kunci 'Sweden' menjadi 'se'
tlds['Sweden'] = 'se'

# f) Menggunakan dictionary comprehension untuk membalik kunci dan nilai
reversed_tlds = {tld: country for country, tld in tlds.items()}
print('f) Kamus yang telah dibalik (TLD - Negara):')
for tld, country in reversed_tlds.items():
    print(f'{tld:<10} - {country}')

# g) Menggunakan dictionary comprehension untuk mengonversi nama negara menjadi huruf besar
uppercase_countries = {tld: country.upper() for tld, country in reversed_tlds.items()}
print('g) Nama negara dalam huruf besar (TLD - Negara):')
for tld, country in uppercase_countries.items():
    print(f'{tld:<10} - {country}')
```

```python
#6.10
# Set pertama
set1 = {'red', 'green', 'blue'}

# Set kedua
set2 = {'cyan', 'green', 'blue', 'magenta', 'red'}

# a) Membandingkan set menggunakan operator perbandingan
# Memeriksa apakah set1 sama dengan set2
is_equal = set1 == set2
print(f'a) Set1 sama dengan set2: {is_equal}')

# Memeriksa apakah set1 tidak sama dengan set2
is_not_equal = set1 != set2
print(f'a) Set1 tidak sama dengan set2: {is_not_equal}')

# Memeriksa apakah set1 adalah subset dari set2
is_subset = set1.issubset(set2)
print(f'a) Set1 adalah subset dari set2: {is_subset}')

# Memeriksa apakah set1 adalah superset dari set2
is_superset = set1.issuperset(set2)
print(f'a) Set1 adalah superset dari set2: {is_superset}')

# b) Menggabungkan set menggunakan operator matematika
# Menggabungkan set1 dan set2 (union)
union_set = set1.union(set2)
print(f'b) Gabungan set1 dan set2: {union_set}')

# Menggabungkan set1 dan set2 dengan operator '|'
or_operator = set1 | set2
print(f"b) Gabungan set1 dan set2 dengan operator '|': {or_operator}")


# Menghitung irisan set1 dan set2 (intersection)
intersection_set = set1.intersection(set2)
print(f'b) Irisan set1 dan set2: {intersection_set}')

# Menghitung irisan set1 dan set2 dengan operator '&'
and_operator = set1 & set2
print(f"b) Irisan set1 dan set2 dengan operator '&': {and_operator}")

# Menghitung selisih set1 dan set2 (difference)
difference_set = set1.difference(set2)
print(f'b) Selisih set1 dan set2: {difference_set}' if difference_set else "b) Selisih set1 dan set2: Tidak ada selisih antara set1 dan set2")


# Menghitung selisih set1 dan set2 dengan operator '-'
subtraction_operator = set1 - set2
print(f"b) Selisih set1 dan set2 dengan operator '-': {subtraction_operator}" if subtraction_operator else "b) Selisih set1 dan set2 dengan operator '-': Tidak ada selisih antara set1 dan set2")
```
