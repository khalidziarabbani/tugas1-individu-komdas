```python
#6.5
def count_duplicate_words(sentence):
    # Mengubah semua huruf menjadi huruf kecil dan memisahkan kalimat menjadi kata-kata
    words = sentence.lower().split()

    # Membuat kamus untuk menghitung jumlah kata
    word_counts = {}

    # Menghitung jumlah kemunculan setiap kata dalam kalimat
    for word in words:
        if word in word_counts:
            word_counts[word] += 1
        else:
            word_counts[word] = 1

    # Mencetak kata-kata yang memiliki lebih dari satu kemunculan (duplikat)
    print("Duplicate Words:")
    for word, count in word_counts.items():
        if count > 1:
            print(f"{word}: {count}")

# Contoh penggunaan skrip
sentence = "This is a sample sentence. This is another sample sentence with sample words."

count_duplicate_words(sentence)
```

```python
#6.6
import string

def get_unique_words(word_list):
    # Menggabungkan semua kata menjadi satu string, mengubah huruf menjadi huruf kecil, dan memisahkan kata-kata
    words = ' '.join(word_list).lower().split()

    # Menghapus tanda baca dari kata-kata
    words = [''.join(char for char in word if char not in string.punctuation) for word in words]

    # Menggunakan set untuk mendapatkan kata-kata unik
    unique_words = set(words)

    # Mengurutkan kata-kata unik dalam urutan alfabet
    unique_words_sorted = sorted(unique_words)

    return unique_words_sorted

# Contoh penggunaan fungsi
sentences = [
    "This is a sample sentence.",
    "This is another sample sentence with more words.",
    "These sentences contain duplicate words, sentences have words."
]

for sentence in sentences:
    word_list = sentence.split()
    unique_words = get_unique_words(word_list)
    print(f"- Kata unik dari '{sentence}': {', '.join(unique_words)}")
```

```python
#6.7
# Meminta pengguna memasukkan sebuah kalimat
sentence = input("Masukkan sebuah kalimat: ")

# Menginisialisasi kamus untuk menghitung jumlah kemunculan huruf
letter_counts = {}

# Menghitung jumlah kemunculan setiap huruf dalam kalimat
for char in sentence:
    if char.isalpha():  # Hanya memproses huruf (mengabaikan spasi dan karakter lainnya)
        char = char.lower()  # Mengabaikan huruf besar/kecil dengan mengubah menjadi huruf kecil
        if char in letter_counts:
            letter_counts[char] += 1
        else:
            letter_counts[char] = 1

# Menampilkan tabel huruf dan jumlah kemunculannya dalam urutan terurut
print("Huruf".ljust(6)+"Jumlah Kemunculan".rjust(20))
for char in sorted(letter_counts.keys()):
    print(char.ljust(6)+letter_counts[char].__str__().center(20))

# Menemukan huruf-huruf yang tidak muncul dalam kalimat
alphabet = set("abcdefghijklmnopqrstuvwxyz")
missing_letters = alphabet - set(letter_counts.keys())

# Menampilkan huruf-huruf yang tidak muncul dalam kalimat
print("Huruf yang tidak muncul dalam kalimat:", ', '.join(sorted(missing_letters)))
```

```python
#6.8
# Kamus yang menghubungkan angka dengan kata-kata
number_words = {
    0: 'ZERO',
    1: 'ONE',
    2: 'TWO',
    3: 'THREE',
    4: 'FOUR',
    5: 'FIVE',
    6: 'SIX',
    7: 'SEVEN',
    8: 'EIGHT',
    9: 'NINE',
    10: 'TEN',
    11: 'ELEVEN',
    12: 'TWELVE',
    13: 'THIRTEEN',
    14: 'FOURTEEN',
    15: 'FIFTEEN',
    16: 'SIXTEEN',
    17: 'SEVENTEEN',
    18: 'EIGHTEEN',
    19: 'NINETEEN',
    20: 'TWENTY',
    30: 'THIRTY',
    40: 'FORTY',
    50: 'FIFTY',
    60: 'SIXTY',
    70: 'SEVENTY',
    80: 'EIGHTY',
    90: 'NINETY'
}

# Meminta pengguna memasukkan jumlah numerik
amount = float(input("Masukkan jumlah numerik (kurang dari 1000): "))

# Fungsi untuk mengonversi dua digit terakhir menjadi kata-kata
def convert_two_digits(number):
    if number in number_words:
        return number_words[number]
    else:
        tens = number // 10 * 10
        ones = number % 10
        return number_words[tens] + ' ' + number_words[ones]

# Mengonversi jumlah numerik menjadi kata-kata
def convert_to_words(amount):
    if amount < 1:
        return "ZERO"
    
    integer_part = int(amount)
    decimal_part = int((amount - integer_part) * 100)

    words = []
    
    if integer_part >= 100:
        hundreds = integer_part // 100
        words.append(number_words[hundreds] + ' HUNDRED')
        integer_part %= 100
    
    if integer_part >= 20:
        tens = integer_part // 10 * 10
        words.append(number_words[tens])
        integer_part %= 10

    if integer_part > 0:
        words.append(number_words[integer_part])

    words.append('AND')
    words.append(f'{decimal_part:02d}/100')

    return ' '.join(words)

# Mengonversi dan menampilkan jumlah numerik dalam kata-kata
amount_in_words = convert_to_words(amount)
print("Jumlah dalam kata-kata:", amount_in_words)
```

```python
#6.9
# Kamus yang menghubungkan negara dengan top-level domains (TLDs)
tlds = {'Canada': 'ca', 'United States': 'us', 'Mexico': 'mx'}

# a) Mengecek apakah kamus mengandung kunci 'Canada'
contains_canada = 'Canada' in tlds
print(f'a) Kamus mengandung kunci "Canada": {contains_canada}')

# b) Mengecek apakah kamus mengandung kunci 'France'
contains_france = 'France' in tlds
print(f'b) Kamus mengandung kunci "France": {contains_france}')

# c) Iterasi melalui pasangan kunci-nilai dan menampilkannya dalam format dua kolom
print('c) Isi kamus (kunci - nilai):')
for country, tld in tlds.items():
    print(f'{country:<15} - {tld}')

# d) Menambahkan pasangan kunci-nilai 'Sweden' dan 'sw' (yang salah)
tlds['Sweden'] = 'sw'

# e) Memperbarui nilai untuk kunci 'Sweden' menjadi 'se'
tlds['Sweden'] = 'se'

# f) Menggunakan dictionary comprehension untuk membalik kunci dan nilai
reversed_tlds = {tld: country for country, tld in tlds.items()}
print('f) Kamus yang telah dibalik (TLD - Negara):')
for tld, country in reversed_tlds.items():
    print(f'{tld:<10} - {country}')

# g) Menggunakan dictionary comprehension untuk mengonversi nama negara menjadi huruf besar
uppercase_countries = {tld: country.upper() for tld, country in reversed_tlds.items()}
print('g) Nama negara dalam huruf besar (TLD - Negara):')
for tld, country in uppercase_countries.items():
    print(f'{tld:<10} - {country}')
```

```python
#6.10
# Set pertama
set1 = {'red', 'green', 'blue'}

# Set kedua
set2 = {'cyan', 'green', 'blue', 'magenta', 'red'}

# a) Membandingkan set menggunakan operator perbandingan
# Memeriksa apakah set1 sama dengan set2
is_equal = set1 == set2
print(f'a) Set1 sama dengan set2: {is_equal}')

# Memeriksa apakah set1 tidak sama dengan set2
is_not_equal = set1 != set2
print(f'a) Set1 tidak sama dengan set2: {is_not_equal}')

# Memeriksa apakah set1 adalah subset dari set2
is_subset = set1.issubset(set2)
print(f'a) Set1 adalah subset dari set2: {is_subset}')

# Memeriksa apakah set1 adalah superset dari set2
is_superset = set1.issuperset(set2)
print(f'a) Set1 adalah superset dari set2: {is_superset}')

# b) Menggabungkan set menggunakan operator matematika
# Menggabungkan set1 dan set2 (union)
union_set = set1.union(set2)
print(f'b) Gabungan set1 dan set2: {union_set}')

# Menggabungkan set1 dan set2 dengan operator '|'
or_operator = set1 | set2
print(f"b) Gabungan set1 dan set2 dengan operator '|': {or_operator}")


# Menghitung irisan set1 dan set2 (intersection)
intersection_set = set1.intersection(set2)
print(f'b) Irisan set1 dan set2: {intersection_set}')

# Menghitung irisan set1 dan set2 dengan operator '&'
and_operator = set1 & set2
print(f"b) Irisan set1 dan set2 dengan operator '&': {and_operator}")

# Menghitung selisih set1 dan set2 (difference)
difference_set = set1.difference(set2)
print(f'b) Selisih set1 dan set2: {difference_set}' if difference_set else "b) Selisih set1 dan set2: Tidak ada selisih antara set1 dan set2")


# Menghitung selisih set1 dan set2 dengan operator '-'
subtraction_operator = set1 - set2
print(f"b) Selisih set1 dan set2 dengan operator '-': {subtraction_operator}" if subtraction_operator else "b) Selisih set1 dan set2 dengan operator '-': Tidak ada selisih antara set1 dan set2")
```
